##### Chap 1.4.5 Morphological Transformations

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 이미지 불러오기 (0은 그레이스케일로 불러옴)
img = cv2.imread('j.png', 0)

# 이미지 출력 (matplotlib를 사용하여 출력)
plt.imshow(img, cmap='gray')  # 이미지를 그레이스케일로 표시
plt.axis('off')  # 축 제거
plt.show()
```

```python
# 커널 생성 (5x5 크기의 1로 이루어진 행렬)
kernel = np.ones((5,5), np.uint8)

# 침식(erosion) 연산 적용, 반복 횟수는 1회
erosion = cv2.erode(img, kernel, iterations=1)

# 이미지 출력 (matplotlib를 사용하여 출력)
plt.imshow(erosion, cmap='gray')  # 이미지를 그레이스케일로 표시
plt.axis('off')  # 축 제거
plt.show()
```

```python
# 커널 생성 (5x5 크기의 1로 이루어진 행렬)
kernel = np.ones((5,5), np.uint8)

erosion = cv2.dilate(img, kernel, iterations=1)

# 이미지 출력 (matplotlib를 사용하여 출력)
plt.imshow(erosion, cmap='gray')  # 이미지를 그레이스케일로 표시
plt.axis('off')  # 축 제거
plt.show()
```

```python
img = cv2.imread('noise.png', 0)

# 커널 생성 (5x5 크기의 1로 이루어진 행렬)
kernel = np.ones((5,5), np.uint8)

opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

# 이미지 출력 (matplotlib를 사용하여 출력)
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(opening, cmap='gray')
plt.title('Opening (Noise Removed)')
plt.axis('off')
```

```python
img = cv2.imread('dot.png', 0)

# 커널 생성 (5x5 크기의 1로 이루어진 행렬)
kernel = np.ones((5,5), np.uint8)

closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

# 이미지 출력 (matplotlib를 사용하여 출력)
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(closing, cmap='gray')
plt.title('Closing (Dot Removed)')
plt.axis('off')
```

```python
img = cv2.imread('j.png', 0)

# 커널 생성 (5x5 크기의 1로 이루어진 행렬)
kernel = np.ones((5,5), np.uint8)

gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

# 이미지 출력 (matplotlib를 사용하여 출력)
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gradient, cmap='gray')
plt.title('gradient')
plt.axis('off')
```

```python
img = cv2.imread('j.png', 0)

# 커널 생성 (5x5 크기의 1로 이루어진 행렬)
kernel = np.ones((5,5), np.uint8)

# 이미지 대비 조정 (명암비 높이기)
img_contrast = cv2.convertScaleAbs(img, alpha=2.0, beta=-100)

# 커널 생성 (9x9 크기의 커널 사용)
kernel = np.ones((9,9), np.uint8)

# Top-hat 모폴로지 연산 적용
tophat = cv2.morphologyEx(img_contrast, cv2.MORPH_TOPHAT, kernel)

# 원본 이미지와 Top-hat 결과 이미지를 나란히 출력
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Contrast Adjusted')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(tophat, cmap='gray')
plt.title('Top-hat')
plt.axis('off')

plt.show()
```

```python
img = cv2.imread('j.png', 0)

# 이미지 대비 조정 (명암비 높이기)
img_contrast = cv2.convertScaleAbs(img, alpha=2.0, beta=-100)

# 커널 생성 (9x9 크기의 커널 사용)
kernel = np.ones((9,9), np.uint8)

blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)

# 이미지 출력 (matplotlib를 사용하여 출력)
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blackhat, cmap='gray')
plt.title('blackhat')
plt.axis('off')
```
