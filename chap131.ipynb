#### chap 1.3.1 Basic Operations on Image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('OIP.jpg')
```

```python
img[100, 100]
```

```python
img[100, 100, 0] = 0
```

```python
img[100, 100]
```

```python
img[100, 100] = [0, 0, 0]
```

```python
img[:10, :10]
```

```python
img[0:50, 0:10] = [0, 0, 0]
```

```python
img[50:100, 50:80] = [50, 50, 50]
```

```python
plt.imshow(img, cmap= 'gray', interpolation= 'bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```
