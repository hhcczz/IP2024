##### Chap 1.4.3 Geometric Transformations of Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('OIP.jpg')
res = cv2.resize(img, None, fx = 30, fy = 30, interpolation = cv2.INTER_CUBIC)
```

```python
cv2.namedWindow('image')

while(1):
    cv2.imshow('image', res)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
    
cv2.destroyAllWindows()
```

```python
img = cv2.imread('OIP.jpg')
rows, cols = img.shape[:2]
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img, M, (cols, rows))
```

```python
cv2.imshow('img', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Rotate

```python
img = cv2.imread('OIP.jpg')
rows, cols = img.shape[:2]
M = cv2.getRotationMatrix2D((cols / 4, rows / 4), 45, 1)
dst = cv2.warpAffine(img, M, (cols, rows))
```

```python
cv2.imshow('img', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Affine Transformation

```python
img = cv2.imread('OIP.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

##### Perspective Transformation

```python
img = cv2.imread('notebook.JPG')
rows,cols,ch = img.shape
pts1 = np.float32([[170,190],[760,196],[808,1096],[100,1080]])
pts2 = np.float32([[0,0],[400,0],[400,600],[0,600]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(400,600))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('menu.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[180,240],[820,220],[820,520],[180,500]])
pts2 = np.float32([[0,0],[820,0],[820,520],[0,520]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(820,500))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
